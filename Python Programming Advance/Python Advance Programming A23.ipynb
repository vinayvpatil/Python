### 1. Create a class Sudoku that takes a string as an argument. The string will contain the numbers of a regular 9x9 sudoku board left to right and top to bottom, with zeros filling up the empty cells.
Attributes  <br>
An instance of the class Sudoku will have one attribute: <br>
- board: a list representing the board, with sublits for each row, with the numbers as integers. Empty cell represented with 0. <br>
Methods <br>
An instance of the class Sudoku wil have three methods: <br>
- get_row(n): will return the row in position n. <br>
- get_col(n): will return the column in position n. <br>
- get_sqr([n, m]): will return the square in position n if only one argument is given, and the square to which the cell in position (n, m) belongs to if two arguments are given. <br>
Example <br>
 <br>
game = Sudoku("417950030000000700060007000050009106800600000000003400900005000000430000200701580")
game.board  [ <br>
  [4, 1, 7, 9, 5, 0, 0, 3, 0], <br>
  [0, 0, 0, 0, 0, 0, 7, 0, 0], <br>
  [0, 6, 0, 0, 0, 7, 0, 0, 0], <br>
  [0, 5, 0, 0, 0, 9, 1, 0, 6], <br>
  [8, 0, 0, 6, 0, 0, 0, 0, 0], <br>
  [0, 0, 0, 0, 0, 3, 4, 0, 0], <br>
  [9, 0, 0, 0, 0, 5, 0, 0, 0], <br>
  [0, 0, 0, 4, 3, 0, 0, 0, 0], <br>
  [2, 0, 0, 7, 0, 1, 5, 8, 0] <br>
] <br>
game.get_row(0)  [4, 1, 7, 9, 5, 0, 0, 3, 0] <br>
game.get_col(8)  [0, 0, 0, 6, 0, 0, 0, 0, 0] <br>
game.get_sqr(1)  [9, 5, 0, 0, 0, 0, 0, 0, 7] <br>
game.get_sqr(1, 8)  [0, 3, 0, 7, 0, 0, 0, 0, 0] <br>
game.get_sqr(8, 3)  [0, 0, 5, 4, 3, 0, 7, 0, 1] <br>
#### Answer :

```python
class Sudoku: 
    def __init__(self, string1):
        self.board = []
        index = 0
        counter = 0
        for i in range(9):
            self.board.append([])
            for j in range(9):
                self.board[index].append(int(string1[counter]))
                counter += 1
            index += 1
        
    def get_row(self, row):
        return self.board[row]
    
    def get_col(self, column):
        return [self.board[i][column] for i in range(9)]
    
    def get_sqr(self, n, *m):
        start_row = (n // 3) * 3
        if m:
            start_column = (m[0] // 3) * 3
        else:
            start_column = 3
        
        output = []
        for i in range(start_row, start_row + 3):
            for j in range(start_column, start_column + 3):
                output.append(self.board[i][j])
        return output
        
game = Sudoku("417950030000000700060007000050009106800600000000003400900005000000430000200701580")
print(game.get_row(0))
print(game.get_col(8))
print(game.get_sqr(1))
print(game.get_sqr(1, 8))
print(game.get_sqr(8,3))
```

#### Answer :

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        self.num_list = []
        self.num_list.append(val)       
    def add_data(self,list1):
        self.num_list.extend(list1)
    def get_data(self):
        return self.num_list
    
class add_two_numbers:
    def __init__(self,x,y):
        self.x = x
        self.y = y 
    def get_data(self):
        it1 = ''.join([str(i) for i in self.x.get_data()[::-1]])
        it2 = ''.join([str(i) for i in self.y.get_data()[::-1]])
        return [int(j) for j in str(int(it1)+int(it2))[::-1]]   
    
lt1 = ListNode(2)
lt1.add_data([4, 3])

lt2 = ListNode(5)
lt2.add_data([6, 4])

print(lt1.get_data()) 
print(lt2.get_data()) 
print(add_two_numbers(lt1, lt2).get_data(),end='\n\n')

lt1 = ListNode(0)
lt2 = ListNode(0)

print(lt1.get_data()) 
print(lt2.get_data()) 
print(add_two_numbers(lt1, lt2).get_data(),end='\n\n')

lt1 = ListNode(9)
lt1.add_data([9,9,9,9,9,9])

lt2 = ListNode(9)
lt2.add_data([9,9,9])

print(lt1.get_data()) 
print(lt2.get_data()) 
print(add_two_numbers(lt1, lt2).get_data())
```

### 3. Write a class called CoffeeShop, which has three instance variables:
1. name : a string (basically, of the shop) <br>
2. menu : a list of items (of dict type), with each item containing the item (name of the item), type (whether a food or a drink) and price. <br>
3. orders : an empty list <br>
and seven methods: <br>
1. add_order: adds the name of the item to the end of the orders list if it exists on the menu, otherwise, return "This item is currently unavailable!" <br>
2. fulfill_order: if the orders list is not empty, return "The {item} is ready!". If the orders list is empty, return "All orders have been fulfilled!" <br>
3. list_orders: returns the item names of the orders taken, otherwise, an empty list. <br>
4. due_amount: returns the total amount due for the orders taken. <br>
5. cheapest_item: returns the name of the cheapest item on the menu. <br>
6. drinks_only: returns only the item names of type drink from the menu. <br>
7. food_only: returns only the item names of type food from the menu. <br>
IMPORTANT: Orders are fulfilled in a FIFO (first-in, first-out) order. <br>
Examples <br>
tcs.add_order("hot cocoa")  "This item is currently unavailable!" <br>
\# Tesha's coffee shop does not sell hot cocoa <br>
tcs.add_order("iced tea")  "This item is currently unavailable!" <br>
\# specifying the variant of "iced tea" will help the process <br>
tcs.add_order("cinnamon roll")   "Order added!" <br>
tcs.add_order("iced coffee")  "Order added!" <br>
tcs.list_orders  ["cinnamon roll", "iced coffee"] <br>
\# all items of the current order <br>
tcs.due_amount()  2.17 <br>
tcs.fulfill_order()  "The cinnamon roll is ready!" <br>
tcs.fulfill_order()  "The iced coffee is ready!" <br>
tcs.fulfill_order()  "All orders have been fulfilled!" <br>
\# all orders have been presumably served <br>
tcs.list_orders()  [] <br>
\# an empty list is returned if all orders have been exhausted <br>
tcs.due_amount()  0.0 <br>
\# no new orders taken, expect a zero payable <br>
tcs.cheapest_item()  "lemonade" <br>
tcs.drinks_only()  ["orange juice", "lemonade", "cranberry juice", "pineapple juice", "lemon iced tea", "vanilla chai latte", "hot chocolate", "iced coffee"] <br>
tcs.food_only()  ["tuna sandwich", "ham and cheese sandwich", "bacon and egg", "steak", "hamburger", "cinnamon roll"] <br>
#### Answer :

```python
Menu = [
    {'name':'Orange Juice','type':'drink','price':5},
    {'name':'Lemonade','type':'drink','price':4},
    {'name':'Cranberry Juice','type':'drink','price':6},
    {'name':'Pineapple Juice','type':'drink','price':6},
    {'name':'Lemon Iced Tea','type':'drink','price':7},
    {'name':'Vanilla Chai Latte','type':'drink','price':8},
    {'name':'Hot Choclate','type':'drink','price':10},
    {'name':'Iced Cofee','type':'drink','price':8},
    {'name':'Tuna Sandwich','type':'food','price':14},
    {'name':'Ham Cheese Sandwich','type':'food','price':16},
    {'name':'Bacon And Egg','type':'food','price': 15},
    {'name':'Chicken Biryani','type':'food','price':40},
    {'name':'Veg Burger','type':'food','price':18},
    {'name':'Cinnamon Roll','type':'food','price':19}
]

class Cofeeshop:
    def __init__(self,name,menu,orders):
        self.name = name
        self.menu = menu
        self.orders = orders
    
    def add_order(self, order_name):
        available = [item['name'].lower() for item in self.menu]
        if order_name in available:
            output = "Order added!"
            self.orders.append(order_name)
        else:
            output = "This item is currently unavailable!"
        return output
    
    def list_orders(self):
        return self.orders
    
    def due_amount(self):
        output = sum([i['price'] for i in self.menu if i['name'].lower() in self.orders])
        return output
    
    def fulfill_order(self):
        if len(self.orders) > 0:
            output = ("The " + self.orders.pop(0) + " is ready!")
        else:
            output = 'All orders have been fulfilled!'
        return output
        
    def cheapest_item(self):
        lowest_price = min([item['price'] for item in self.menu])
        output = [i['name'] for i in self.menu if i['price'] == lowest_price]
        return output[0]
        
    def drinks_only(self):
        output = [i['name'] for i in self.menu if i['type'] == 'drink']
        return output
    
    def food_only(self):
        output = [i['name'] for i in self.menu if i['type'] == 'food']
        return output
    
tcs = Cofeeshop('Tesha\'s Cofee Shop',Menu,[])
print(tcs.add_order("hot cocoa"))
print(tcs.add_order("iced tea"))
print(tcs.add_order("cinnamon roll"))
print(tcs.add_order("iced cofee"))
print(tcs.list_orders())
print(tcs.due_amount())
print(tcs.fulfill_order())
print(tcs.fulfill_order())
print(tcs.fulfill_order())
print(tcs.list_orders())
print(tcs.due_amount())
print(tcs.cheapest_item())
print(tcs.drinks_only())
print(tcs.food_only())
```

### 4. In this challenge, write a funcion loneliest_number to find the last Lonely number inside a sequence. A number is Lonely if the distance from its closest Prime sets a new record of the sequence.
Sequence = from 0 to 3 <br>
\# Any number lower than 3 doesn't have a Prime preceeding it... <br>
\# ...so that you'll consider only its next closest Prime. <br>
0 has distance 2 from its closest Prime (2) <br>
\# It's a new record! 0 It's the first lonely number of the sequence <br>
1 has distance 1 from its closest Prime (2) <br>
2 has distance 1 from 3 <br>
3 has distance 1 from 2 <br>
\# The sequence 0 to 3  has only one Lonely number: 0 <br>
Sequence = Numbers from 5 to 10 <br>
5 has distance 2 from its closest Prime (3 or 7) <br>
\# It's a new record! 5 It's the first lonely number of the sequence <br>
6 has distance 1 from 5 or 7 <br>
7 has distance 2 from 5 <br>
8 has distance 1 from 7 <br>
9 has distance 2 from 7 or 11 <br>
10 has distance 1 from 11 <br>
\# The sequence 5 to 10  has only one Lonely number: 5 <br>
Sequence = Numbers from 19 to 24 <br>
19 has distance 2 from its closest Prime (17) <br>
\# It's a new record! 19 It's the first lonely number of the sequence <br>
20 has distance 1 from 19 <br>
21 has distance 2 from 5 <br>
22 has distance 1 from 23 <br>
23 has distance 4 from 17 or 29 <br>
\# It's a new record! 23 is the second lonely number of the sequence <br>
24 has distance 1 from 23 <br>
\# The sequence 19 to 24  has two Lonely numbers: 19 and 23 <br>
The function loneliest_number must accept two integers lo and hi being the inclusive bounds of the sequence to analyze, and returns a dictionay (dict) object with the following keys and values: <br>
- number: is the last Lonely number found in the given sequence; <br>
- distance: is the distance of the number from its closest Prime; <br>
- closest: is the Prime closest to number (if two Primes are equally distant from number, return the higher Prime). <br>
Examples <br>
loneliest_number(0, 22)  { <br>
  number: 0, distance: 2, closest: 2 <br>
} <br>
loneliest_number(8, 123)  { <br>
  number: 53, distance: 6, closest: 59 <br>
} <br>
loneliest_number(938, 1190)  { <br>
  number: 1140, distance: 11, closest: 1151 <br>
} <br>
loneliest_number(120, 1190)  { <br>
  number: 211, distance: 12, closest: 223 <br>
} <br>
#### Answer :

```python
from math import floor, sqrt
def find_prev_prime(n):
    i = 2
    while i != n:
        if not n % i:
            n -= 1
            i = 2
        else:
            i += 1
    return n

def find_next_prime(n):
    i = 2
    while i < floor(sqrt(n)) + 1:
        if not n % i:
            n += 1
            i = 2
        else:
            i += 1
    return n

def loneliest_number(lower, upper):
    output = {
        'number' : 0,
        'distance' : None,
        'closest' : None
    }
    
    if lower > 1:
        prev_prime = find_prev_prime(lower - 1)
        next_prime = find_next_prime(lower + 1)
    else:
        prev_prime = 2
        next_prime = 3

    temp_prime = None
    for i in range(lower, upper + 1):
        
        if next_prime == i:
            temp_prime = next_prime
            next_prime = find_next_prime(i + 1)
        
        distance = min(abs(i - prev_prime) , abs(i - next_prime))
        #abs() function return the absolute value.

        if output['distance'] == None:
            output['distance'] = distance
            output['number'] = i
            if abs(i - prev_prime) > abs(i - next_prime):
                output['closest'] = next_prime
            else:
                output['closest'] = prev_prime
                
        else:
            if distance > output['distance']:
                output['distance'] = distance
                output['number'] = i
                if abs(i - prev_prime) >= abs(i - next_prime):
                    output['closest'] = next_prime
                else:
                    output['closest'] = prev_prime        
        
        if temp_prime == i:
            prev_prime = temp_prime
        
    return output
    
print(loneliest_number(0, 22))
print(loneliest_number(8, 123))
print(loneliest_number(938, 1190))
print(loneliest_number(120, 1190))
```

### 5. Implement a class Selfie that can store the current state of the object in the form of binary string. It can take multiple pictures and then recover to a state it was before. During testing an object will be provided with new attributes and their values. It will store its state. Then the values will be changed. Then it will be given new attributes. It will store its state again. It will be repeated few times.
Later the states of the object will be recovered given an index. The return value should be a new Selfie with the requested historic state and the state history of the new object should be updated with a copy of current object's state history. <br>
The object also knows how many states it has stored. If the index is not within the range of stored states, the object stays as is. If the argument is invalid, n < 0 or n >= self.n_states(), the current object (or a copy thereof) should be returned. <br>
Examples  <br>
p = Selfie() <br>
p.x = 2 <br>
p.save_state() <br>
p.x = 5 <br>
p = p.recover_state(0) <br>
p.x  2 <br>
#### Answer :

```python
class Selfie:
    def __init__(self):
        self.x= 2
        a=[]
        a.append(self.x)
        self.a = a
        
    def save_state(self):
        self.x=5
        self.a.append(self.x)
    
    def recover_state(self,n):
        try:
            self.x = self.a[n]
        except:
            self.x
p = Selfie()
print(p.x) 
p.save_state()
print(p.x )
p.recover_state(0)
print(p.x)    
```
